```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def read_csv(csv_path):
    np_path_XYs = np.genfromtxt(csv_path , delimiter=',') 
    path_XYs = []
    for i in np.unique(np_path_XYs[:, 0]):
        npXYs = np_path_XYs[np_path_XYs[:, 0] == i][:, 1:] 
        XYs = []
        for j in np.unique(npXYs[:, 0]):
            XY = npXYs[npXYs[:, 0] == j][:, 1:]
            XYs.append(XY) 
        path_XYs.append(XYs)
    return path_XYs
```

```python
colours = ['k', 'y', 'r', 'b', 'g', 'c', 'm']
def plot(path_XYs,save_path):
    fig, ax = plt.subplots(tight_layout=True, figsize=(8, 8))
    for i, XYs in enumerate(path_XYs): 
        c = colours[i % len(colours)] 
        for XY in XYs:
            ax.plot(XY[:, 0], XY[:, 1], c=c, linewidth=2) 
    ax.set_aspect('equal')
#     ax.axis('off')
    plt.savefig(save_path, bbox_inches='tight', pad_inches=0, transparent=True)
    plt.show()
```

```python
def regularize_doodles(path_XYs):
    regularized_path_XYs = []
    
    for polyline in path_XYs:
        # Convert polyline to a numpy array of float32 type
        contour = np.array(polyline, dtype=np.float32)
        
        # Apply Gaussian blur to smooth the contour
        contour_blurred = cv2.GaussianBlur(contour, (5, 5), 0)
        
        # Approximate the contour with fewer points
        epsilon = 0.01 * cv2.arcLength(contour_blurred, True)
        regularized_contour = cv2.approxPolyDP(contour_blurred, epsilon, True)
        regularized_contour = regularized_contour.reshape(-1, 2)
        # Append the regularized contour as a numpy array to the list
        regularized_contour = regularized_contour[np.newaxis, :, :]
        regularized_path_XYs.append(regularized_contour)
    
    return regularized_path_XYs
```

```python
import cv2
class ShapeDetector:
    def __init__(self):
        pass
    def detect(self, c):
        # initialize the shape name and approximate the contour
        shape = "unidentified"
        peri = cv2.arcLength(c, True)
        approx = cv2.approxPolyDP(c, 0.04 * peri, True)
        # if the shape is a triangle, it will have 3 vertices
        if len(approx) == 3:
            shape = "triangle"
        # if the shape has 4 vertices, it is either a square or
        # a rectangle
        elif len(approx) == 4:
            # compute the bounding box of the contour and use the
            # bounding box to compute the aspect ratio
            (x, y, w, h) = cv2.boundingRect(approx)
            ar = w / float(h)
            # a square will have an aspect ratio that is approximately
            # equal to one, otherwise, the shape is a rectangle
            shape = "square" if ar >= 0.95 and ar <= 1.05 else "rectangle"
        # if the shape is a pentagon, it will have 5 vertices
        elif len(approx) == 5:
            shape = "pentagon"
            # otherwise, we assume the shape is a circle
        else:
            shape = "circle"
        # return the name of the shape
        return shape
```

## Symmetry: Vertical and Horizonal

```python
img = cv2.imread("image.png")

# check is image is loaded or not
if img is None:
    print("Error to open image file")
    
else:
    # converting image to gray scale image
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
   
    # draw a vertical line to visualize symmetry
    img_symmetry = img.copy()
    
    cv2.line(img_symmetry,(gray.shape[1] //2,0),(gray.shape[1] //2,gray.shape[0]),(0,0,255),5)
    
#   display the original and symmetry visualizatoin
    cv2.imshow('Original',img)
    cv2.imshow('Vertical Flip',img_symmetry)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
